```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips)         #
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips, hue = 'smoker')   # its a smoker column 
```

```python
tips
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips, hue = 'day')       #give you the day column value
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips, hue = 'day', style= 'smoker') #grouping for day and smoker diffrences 
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips, hue = 'day', style= 'smoker',size='size')   #show dya size smoker 
```

```python
sns.relplot(x = 'tip',  y = 'total_bill', data = tips, size='tip')
```
